```python
%matplotlib inline
fgr=plt.figure()
a=np.random.uniform(0,100,100)
b=np.random.uniform(0,100,100)
ax1=fgr.add_axes([0.1,0.1,0.8,0.8])
ax1.scatter(a,b,color="r")
ax2=fgr.add_axes([0.5,0.5,0.4,0.4])
ax2.scatter(a**2,b**2,alpha=0.5)
ax2.patch.set_alpha(0.5)
```

```python
fig,ax=plt.subplots()
x=np.arange(10)
y=np.random.randint(1,100,10)
z=np.random.randint(1,100,10)
ax.stackplot(x,[y,z],labels=["y1","z"]);
ax.legend()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
a=np.random.uniform(-5,5,100)
b=np.random.uniform(-5,5,100)
axes[0,0].scatter(a,b)
axes[0,0].set_title("Scatter")
axes[0,0].spines["bottom"].set_color("r")
axes[0,1].hist(a)
axes[0,1].set_title("Histogram")
axes[1,0].plot(a,b)
axes[1,1].pie(a,startangle=0);
```

```python
fig,ax=plt.subplots()
x=np.arange(10)
y1=x**2
y2=x**5

ax.plot(x,y1,"r")
ax.spines["left"].set_color("red")
ex=ax.twinx()
ex.plot(x,y2)


for label in ax.get_yticklabels():
    label.set_color("red")
    label.set_fontsize(12)
ax.spines["bottom"].set_color("red")

for label in ex.get_yticklabels():
    label.set_color("blue")
    label.set_fontsize(12)
```
