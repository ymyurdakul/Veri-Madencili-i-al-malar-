```python
%config IPCompleter.greedy=True
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
dolar_kuru=[1.70,1.90,2.10,2.30]
yıllar=np.array([2012,2013,2014,2015])
asgari=[850,900,1000,1200]
fig,ax=plt.subplots()

ax.plot(yıllar,dolar_kuru)
ax.locator_params(axis="x",integer=True);

for label in ax.get_yticklabels():
    label.set_color("blue")
ex=ax.twinx()
ex.plot(yıllar,asgari,"r")
for label in ex.get_yticklabels():
    label.set_color("red")
ax.legend(["dolar artısı"],loc=0);
ex.legend(["asgari artısı"],loc=1);
ax.set_title("Yıllara Göre Dolar ve Asgari ücretin değişimi");
```
