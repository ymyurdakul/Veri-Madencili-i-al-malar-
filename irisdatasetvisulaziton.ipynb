```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
plt.figure()
colors=["R","G","B"]
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
for i,name in enumerate(iris.target_names):
    plt.scatter(iris.data[i==iris.target,0],iris.data[i==iris.target,1],c=colors[i])
lgnd=plt.legend(iris.target_names)
lgnd.legendHandles[0]._size=[500]
plt.grid()
print(iris.feature_names)
np.corrcoef(iris.data[iris.target==2,0],iris.data[iris.target==2,2])
```
